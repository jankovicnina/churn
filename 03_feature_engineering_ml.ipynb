```python
import pandas as pd
import numpy as np
import pyarrow
```

```python
df = pd.read_parquet('train.parquet', engine = 'pyarrow')
```

```python
def get_registration_features(user_data):
    features = {}
    
    if 'registration' in user_data.columns:
        reg_time = user_data['registration'].iloc[0]
        first_activity = user_data['time'].min()
        
        if pd.notna(reg_time):
            gap_days = (first_activity - reg_time).days
            features['reg_to_first_gap'] = max(gap_days, 0)
            features['reg_gap_week'] = 1 if gap_days >= 7 else 0
            features['reg_gap_month'] = 1 if gap_days >= 30 else 0
    
    return features
```

```python
def get_engagement_features(user_data):
    features = {
        'total_events': len(user_data),
        'total_sessions': user_data['sessionId'].nunique(),
        'song_ratio': (user_data['page'] == 'NextSong').sum() / max(len(user_data), 1),
        'thumbs_up_ratio': (user_data['page'] == 'Thumbs Up').sum() / max(len(user_data), 1),
        'thumbs_down_ratio': (user_data['page'] == 'Thumbs Down').sum() / max(len(user_data), 1),
    }
    return features
```
